```python
# import library
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv')
df = df.copy()
# tampilkan 5 data teratas
df.head()
```

```python
# tampilkan 5 data terbawah
df.tail()
```

```python
# convert tipe data FWI menjadi float
df['FWI'] = pd.to_numeric(df['FWI'], errors='coerce')
```

```python
# cek apakah ada data yang kosong
df.isnull().sum()
```

```python
# buang data yang kosong
df.dropna(inplace=True)
```

```python
# cek lagi apakah ada data yang kosong
df.isnull().sum()
```

```python
# drop atribut year karena semuanya terjadi pada tahun yang sama
# drop atribut classes karena tidak bisa dibandingkan
# df = df.drop('year', axis=1)
df = df.drop(df.columns[[2,13]], axis=1)
```

```python
df.head()
```

```python
# melihat korelasi antar variabel
sns.heatmap(df.corr(), annot=True, cmap='viridis', linewidths=0.5)
```

```python
# HISTOGRAM
df.hist(figsize=(20,14), color='blue')
```

```python
# LINEPLOT
sns.lineplot(x='Temperature', y='day', data=df, color='green')
```

```python
# PAIRPLOT
sns.pairplot(df)
```
